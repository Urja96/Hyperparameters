**Converting the object data to numeric data type**

```python
df_gbm = df_gbm.iloc[:,2:].apply(pd.to_numeric, downcast='integer')
```

```python
df_gbm.dtypes
```

```python
ranges = []
for key,val in df_gbm.dtypes.items():
    print(key,val)
    if val == 'int': 
         ranges.append([key,[df_gbm[key].min(),df_gbm[key].max()]])
    if val == 'float':
         ranges.append([key,[df_gbm[key].min(),df_gbm[key].max()]])
```

### Finding the ranges of the hyper parameter

```python
ranges
```
